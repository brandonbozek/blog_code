# Overview

My previous [two](https://brandonbozek.github.io/honey-part1/) [posts](https://brandonbozek.github.io/honey-part2/) explored US honey production and identified key relationships
between the supply of honey and its price. In this post, I will build on insights
gained there to develop models that will allow me to make predictions
about the future US honey market. In particular, I want to determine: How much honey will
be on the domestic market at some time in the future? What will be the price of honey at that time?

Based on what I learned in my previous projects, the answer to the second question depends on the answer to the first. So, I will need to build two predictive models:
one model to forecast the future market size of honey in each state and another model
to predict, based on the results of the first model, the future price of that honey.

The future market size of honey can be forecasted with a
time-series model. Time-series models assess both long term and short term
trends over the timespan of the dataset to predict future events. Since
each state is independent, the market size of each state will need to be
modeled separately. Therefore, the first model I will build is an ensemble of
time-series models for each state where I determined the optimal forecasting
method to predict the amount of honey that state will produce in subsequent
years.

Next, the price of honey will be modeled using machine learning regression algortihms. Feeding
the results of the market size forecast into this model, I can predict the price of
honey in the future.

The blog [post](https://brandonbozek.github.io/honey-part3/) contains the broad brush description of the construction of these
two models. This notebook contains the code used to make the figures as well as the full procedure for selecting and evaluating the various models. 

```python
#Import relevant libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

#import stats and time series libraries
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt
from sklearn.metrics import mean_squared_error
from pandas.tseries.offsets import DateOffset

#set my plotting environements:
%matplotlib inline
sns.set()
plt.rcParams["patch.force_edgecolor"] = True
sns.set_context("notebook",font_scale=1.25)
```

Import dataset from Part 1:

```python
honey = pd.read_csv('honeyproduction.csv')#honey data from 1998-2012
```

Add a column that enumerates states, which will be useful later when applying predictive models 

```python
lst_state = list(honey['state'].unique())
d = {}
for ix,st in enumerate(lst_state):
    d[st] = ix
honey['stnum'] = honey['state'].map(d)
```

The honey market dataset used in part 1 and 2 covered the years 1999-2012.
Wanting this project to be more current, I downloaded the raw data for US
honey production for 2013-2017 from the [US Department of
Agriculture](https://www.nass.usda.gov/). I cleaned and processed the data below ('honey_late') to
match the format of the earlier dataset above ('honey'). The notebook containing that
procedure can be found [here](https://github.com/brandonbozek/blog_code/blob/master/Honey_Production_Data_Wrangling.ipynb).

```python
honey_late = pd.read_csv('honeyproduction_late.csv')#honey data from 2013-2017
```

```python
honey_late['stnum'] = honey_late['state'].map(d)
```

Combining the data together, I have a lot of latitude in terms of how to
build a predictive model. Ultimately, the model I will build could be used to
predict this year's production and price, however then I would need to wait
to see if I was correct (and that is no fun). So, I chose instead to make a
multiyear prediction for 2015-2017 so I can compare the model predictions
with the actual supply of honey on the US market. This is a little like
pretending I am data scientist in 2014 and predicting the price for the
next 3 years.

```python
honey_tot = pd.concat([honey,honey_late],ignore_index=True)
```

For the full dataset, I create the 'marketsize' column as in Part 2.

```python
honey_tot['stocks_shift'] = honey_tot.groupby('state').stocks.shift(1)
honey_tot[honey_tot['year']>1998]
honey_tot['marketsize']=honey_tot['stocks_shift']+honey_tot['totalprod']
```

```python
honey_tot.tail()
```

Since I am focusing on market size here, I can safely drop all rows from 1998 where there is no market size data.

```python
honey_tot.dropna(axis=0,inplace=True)
```

The first goal is to build a time series model to predict the market size of each
state for 2015-2017 based on the data from 1999-2014. In order to develop some
intuition on how best to do this, I examined several forecasting methods for the
market size on a singe state, California, in detail. California is one of the larger
honey producing states, so it is great sandbox for experimentation.

In order to evaluate the effectiveness of the modelling process, I needed to divide the
California dataset into a training and testing set. I settled on a training
set of the California honey market size for 1999-2011 which I used to build a
model and make a multiyear prediction for 2012-2015. Then, I evaluated the accuracy of those predictions by comparing against the testing
dataset of the acutal California honey market size values for 2012-2015.

```python
CA_train = honey_tot[(honey_tot['state']=='CA') & (honey_tot['year']<2012)][['year','marketsize']]
CA_test = honey_tot[(honey_tot['state']=='CA') & (honey_tot['year']>=2012) & (honey_tot['year']<2015)][['year','marketsize']]
CA_full = honey_tot[(honey_tot['state']=='CA') & (honey_tot['year']<2015)][['year','marketsize']]
```

The time-series analysis tools I will use require the year to be in datetime format. I make that conversion below. 

```python
CA_train['year'] = pd.to_datetime(CA_train['year'],format='%Y')
CA_train.set_index('year',inplace=True)
CA_test['year'] = pd.to_datetime(CA_test['year'],format='%Y')
CA_test.set_index('year',inplace=True)
CA_full['year'] = pd.to_datetime(CA_full['year'],format='%Y')
CA_full.set_index('year',inplace=True)
```

I am considering three time-series models (in order of increasing complexity): a) a
linear trend model, b) a moving average model, and c) a Holt-Winters seasonal
model. The simplest thing to try first is a linear fit to the data to predict the future trend. This may be overly simple, but it is a good place to start.

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(CA_train.index.year.values,CA_train.values.T)
future_dates = [CA_train.index[-1] + DateOffset(years=x) for x in range(1,4)]
CA_pred_lin = pd.DataFrame(index=future_dates[0:],columns=CA_train.columns)
CA_model_lin = pd.DataFrame(index=CA_train.index,columns=CA_train.columns)
CA_model_lin['marketsize'] = slope*np.float32(CA_train.index.year.values) + intercept
CA_pred_lin['marketsize'] = slope*np.float32(CA_pred_lin.index.year.values) + intercept
CA_err_lin = np.sqrt(mean_squared_error(CA_test,CA_pred_lin))
print 'CA linear model error: ',CA_err_lin
```

The plot below shows the linear fit of the training data and the predicted values.

```python
fig, ax = plt.subplots(figsize=(8,6))
CA_full['marketsize'].plot(label='CA Market Size')
CA_model_lin['marketsize'].plot(label='Linear Model')
CA_pred_lin['marketsize'].plot(label='Linear Prediction')
ax.set_ylabel('Market Size (lbs)',fontsize=15)
ax.set_xlabel('Year',fontsize=15)
plt.ticklabel_format(axis='y',useMathText=True)
plt.legend();
plt.savefig('CA_MarketSize_Lin.png');
```

Next, I will consider the moving average model. The moving average model predicts the i-th point by
averaging that step with the previous N-1 steps, where N is a tunable
parameter that is determined through an optimization routine. I am building this model on top of the linear model in order to forecast to future dates. 

```python
CA_roll = pd.concat([CA_train,CA_pred_lin]) #building this model on linear case
order = [2,3,4,5] #trying different window sizes
rmse_arr = []
for j in order:#loop over different model parameterizations
    CA_model_ma_temp = CA_roll.rolling(window=j).mean()
    CA_pred_ma_temp = CA_model_ma_temp.iloc[-3:]
    rmse_arr.append(np.sqrt(mean_squared_error(CA_test,CA_pred_ma_temp)))
indmin = np.argmin(rmse_arr) #determine model with minimal error
print order[indmin],rmse_arr[indmin]
wd = order[indmin]
#forecast production
CA_pred_ma = CA_roll.rolling(window=wd).mean().iloc[-3:]
```

```python
fig, ax = plt.subplots(figsize=(8,6))
CA_full['marketsize'].plot(label='CA Market Size')
CA_full['marketsize'].rolling(window=wd).mean().iloc[:-3].plot(marker='o',label='Moving Average Model')
CA_pred_ma['marketsize'].plot(marker='o',label='Moving Average Prediction')
ax.set_ylabel('Market Size (lbs)',fontsize=15)
ax.set_xlabel('Year',fontsize=15)
plt.ticklabel_format(axis='y',useMathText=True)
plt.legend()
plt.savefig('CA_MarketSize_MovAvg.png')
```

The last model I will consider is Holt-Winters seasonal model. The
Holt-Winters model uses an exponential weighted moving average and adds a
seasonal variation to respectively model long and short term trends in the
data. I did some experimenting with a seasonal Autoregressive Integrated Moving
Average (SARIMA) model (not shown here), which is a popular choice in time-series modeling. But
given the relative small sample of data points, I decided that it was too complex
for this dataset. 

```python
order = [2,3,4,5]
rmse_arr = []
for j in order:#loop over different model parameterizations
    try:
        CA_model_hw_temp = ExponentialSmoothing(CA_train,trend='add',seasonal='add',seasonal_periods=j,freq='AS-JAN').fit()
        CA_pred_hw_temp = CA_model_hw_temp.predict(start=CA_test.index[0], end=CA_test.index[-1])
        rmse_arr.append(np.sqrt(mean_squared_error(CA_test,CA_pred_hw_temp)))
    except:
        print 'H-W fails for order: ',j
indmin = np.argmin(rmse_arr)
print order[indmin],rmse_arr[indmin] #determine model with minimal error
sp = order[indmin]
#build optimal model and forecast production
CA_model_hw = ExponentialSmoothing(CA_train,trend='add',seasonal='add',seasonal_periods=sp,freq='AS-JAN').fit()
CA_pred_hw = CA_model_hw.predict(start=CA_test.index[0], end=CA_test.index[-1])
```

```python
fig, ax = plt.subplots(figsize=(8,6))
CA_full['marketsize'].plot(label='CA Market Size')
CA_model_hw.predict(start=CA_train.index[0], end=CA_train.index[-1]).plot(marker='o',label='Holt-Winters Model')
CA_pred_hw.plot(marker='o',label='Holt-Winters Prediction')
ax.set_ylabel('Market Size (lbs)',fontsize=15)
ax.set_xlabel('Year',fontsize=15)
plt.ticklabel_format(axis='y',useMathText=True)
plt.legend()
plt.savefig('CA_MarketSize_HW.png')
```

Isolating the model predicitons and the test data: 

```python
fig, ax = plt.subplots(figsize=(8,6))
CA_test['marketsize'].plot(label='CA Market Size')
CA_pred_lin['marketsize'].plot(marker='o',label='Linear Prediction')
CA_pred_ma['marketsize'].plot(marker='o',label='Moving Average Prediction')
CA_pred_hw.plot(marker='o',label='Holt-Winters Prediction')
ax.set_ylabel('Market Size (lbs)',fontsize=15)
ax.set_xlabel('Year',fontsize=15)
plt.ticklabel_format(axis='y',useMathText=True)
plt.legend()
plt.savefig('CA_MarketSize_Zoom.png')
```

Now that I have confirmed that these models are competetive in forecasting the production of one state, I will build a pipeline that identifies the best method for each state and forecasts that state's market size for 2015-2017. 

First I define the dataframes that will store the error and predictions for each model. I also remove states from the forecasting model that do not have a non-zero market size for all years in the dataset. These states stopped producing honey at some point in the dataset and my model is not able to account for whether that state will restart production and, if so, when. 

```python
err_df = pd.DataFrame(columns=['state','err_lin','err_ma','err_hw','order_hw','order_ma'])
st_mdl_list = honey_tot.groupby('state').filter(lambda x: x['marketsize'].count() == 19)['state'].unique()
err_df['state'] = st_mdl_list
model_df = pd.DataFrame(columns=['state','year','pred_lin','pred_ma','pred_hw'])
```

Next I use the Holt-Winters method to predict future market size of each state after determining that state's optimal parameterization for this model. I also catalog the error of the best fit model for comparison with other models.

```python
#Holt-Winters Modeling, Prediciton and Error measure
for state in st_mdl_list:
    #set up temporary dataframes
    df_train = honey_tot[(honey_tot['state']==state) & (honey_tot['year']<2012)][['year','marketsize']]
    df_test = honey_tot[(honey_tot['state']==state) & (honey_tot['year']>=2012) & (honey_tot['year']<2015)][['year','marketsize']]
    df_full = honey_tot[(honey_tot['state']==state) & (honey_tot['year']<2015)][['year','marketsize']]
    df_pred = honey_tot[(honey_tot['state']==state) & (honey_tot['year']>=2015)][['year','marketsize']]
    
    df_train['year'] = pd.to_datetime(df_train['year'],format='%Y')
    df_train.set_index('year',inplace=True)
    df_test['year'] = pd.to_datetime(df_test['year'],format='%Y')
    df_test.set_index('year',inplace=True)
    df_full['year'] = pd.to_datetime(df_full['year'],format='%Y')
    df_full.set_index('year',inplace=True)
    df_pred['year'] = pd.to_datetime(df_pred['year'],format='%Y')
    df_pred.set_index('year',inplace=True)
    
    # determine optimal model
    order = [2,3,4,5]
    rmse_arr = []
    for j in order:
        try: 
            mdl = ExponentialSmoothing(df_train,trend='add',seasonal='add',seasonal_periods=j,freq='AS-JAN').fit()
            prdct = mdl.predict(start=df_test.index[0], end=df_test.index[-1])
            rmse_arr.append(np.sqrt(mean_squared_error(df_test,prdct)))
        except:
            rmse_arr.append(1.0e12)#large error set for failure to fit dataset  
    indmin = np.argmin(rmse_arr)
    err_df.loc[(err_df['state']==state),'err_hw'] = rmse_arr[indmin]
    err_df.loc[(err_df['state']==state),'order_hw'] = order[indmin]
    
    #save error and model predictions to relevant dataframes
    spd = order[indmin]
    try:
        mdl = ExponentialSmoothing(df_full,trend='add',seasonal='add',seasonal_periods=spd,freq='AS-JAN').fit()
        prdct = mdl.predict(start=df_pred.index[0], end=df_pred.index[-1])
        if sum(prdct < 0):
            print 'state with predicted negative marketsize: ',state
            prdct[prdct<0] = 0#if statement unnecessary
    except:
        prdct = np.zeros(3)
    prdct_df = pd.DataFrame(prdct,columns=['pred_hw'])
    prdct_df['state'] = state
    prdct_df.reset_index(inplace=True)
    prdct_df.rename(columns = {'index':'year'}, inplace = True)
    model_df = pd.concat([model_df,prdct_df],sort=True,ignore_index=True)
```

The pipeline for the linear model is below:

```python
#linear prediction and error
for state in st_mdl_list:#honey_tot['state'].unique():
    #set up temporary dataframes
    df_train = honey_tot[(honey_tot['state']==state) & (honey_tot['year']<2012)][['year','marketsize']]
    df_test = honey_tot[(honey_tot['state']==state) & (honey_tot['year']>=2012) & (honey_tot['year']<2015)][['year','marketsize']]
    df_train['year'] = pd.to_datetime(df_train['year'],format='%Y')
    df_train.set_index('year',inplace=True)
    df_test['year'] = pd.to_datetime(df_test['year'],format='%Y')
    df_test.set_index('year',inplace=True)
    
    df_full = honey_tot[(honey_tot['state']==state) & (honey_tot['year']<2015)][['year','marketsize']]
    df_full['year'] = pd.to_datetime(df_full['year'],format='%Y')
    df_full.set_index('year',inplace=True)
    
    
    slope, intercept, r_value, p_value, std_err = stats.linregress(df_train.index.year.values,df_train.values.T)
    test_dates = [df_train.index[-1] + DateOffset(years=x) for x in range(1,4)]
    df_pred_lin = pd.DataFrame(index=test_dates[0:],columns=df_train.columns)
    df_pred_lin['marketsize'] = slope*np.float32(df_pred_lin.index.year.values) + intercept
    if sum(df_pred_lin['marketsize'] < 0): #account for incorrect negative values
        print 'state with negative marketsize: ',state
        df_pred_lin.loc[(df_pred_lin['marketsize']<0),'marketsize'] = 0
    errlin = np.sqrt(mean_squared_error(df_test,df_pred_lin))
    
    #save error and model predictions to relevant dataframes
    slp_full, int_full, r_full, p_full, std_err_full = stats.linregress(df_full.index.year.values,df_full.values.T)
    future_dates = [df_full.index[-1] + DateOffset(years=x) for x in range(1,4)]
    df_pred_fut = pd.DataFrame(index=future_dates[0:],columns=df_full.columns)
    df_pred_fut['marketsize'] = slp_full*np.float32(df_pred_fut.index.year.values) + int_full
    if sum(df_pred_fut['marketsize'] < 0):#account for incorrect negative values
        print 'state with predicted negative marketsize: ',state
        df_pred_fut.loc[(df_pred_fut['marketsize']<0),'marketsize'] = 0
    model_df.loc[model_df['state']==state,'pred_lin'] = df_pred_fut['marketsize'].values
    err_df.loc[err_df['state']==state,'err_lin'] = errlin
```

Lastly, I finished the marketsize forecasting model by building the remaining pipeline for a moving average model:

```python
# Moving average error, order
for state in st_mdl_list:
    #set up temporary dataframes
    df_train = honey_tot[(honey_tot['state']==state) & (honey_tot['year']<2012)][['year','marketsize']]
    df_test = honey_tot[(honey_tot['state']==state) & (honey_tot['year']>=2012) & (honey_tot['year']<2015)][['year','marketsize']]
    df_full = honey_tot[(honey_tot['state']==state) & (honey_tot['year']<2015)][['year','marketsize']]
    
    df_train['year'] = pd.to_datetime(df_train['year'],format='%Y')
    df_train.set_index('year',inplace=True)
    df_test['year'] = pd.to_datetime(df_test['year'],format='%Y')
    df_test.set_index('year',inplace=True)
    df_full['year'] = pd.to_datetime(df_full['year'],format='%Y')
    df_full.set_index('year',inplace=True)
    
    # determine optimal model
    slope, intercept, r_value, p_value, std_err = stats.linregress(df_train.index.year.values,df_train.values.T)
    test_dates = [df_train.index[-1] + DateOffset(years=x) for x in range(1,4)]
    df_pred_lin = pd.DataFrame(index=test_dates[0:],columns=df_train.columns)
    df_pred_lin['marketsize'] = slope*np.float32(df_pred_lin.index.year.values) + intercept
    if sum(df_pred_lin['marketsize'] < 0):
        df_pred_lin.loc[(df_pred_lin['marketsize']<0),'marketsize'] = 0
        
    df_roll_mdl = pd.concat([df_train,df_pred_lin])
    order = [2,3,4,5]
    rmse_arr = []
    for j in order:
        df_model = df_roll_mdl.rolling(window=j).mean()
        df_pred = df_model.iloc[-3:]#CAmodel.predict(start=CA_test.index[0], end=CA_test.index[-1])
        rmse_arr.append(np.sqrt(mean_squared_error(df_test,df_pred)))
    
    
    #save error and model predictions to relevant dataframes
    indmin = np.argmin(rmse_arr)
    err_df.loc[(err_df['state']==state),'err_ma'] = rmse_arr[indmin]
    err_df.loc[(err_df['state']==state),'order_ma'] = order[indmin]
    
    slp_full, int_full, r_full, p_full, std_err_full = stats.linregress(df_full.index.year.values,df_full.values.T)
    future_dates = [df_full.index[-1] + DateOffset(years=x) for x in range(1,4)]
    df_pred_fut = pd.DataFrame(index=future_dates[0:],columns=df_full.columns)
    df_pred_fut['marketsize'] = slp_full*np.float32(df_pred_fut.index.year.values) + int_full
    if sum(df_pred_fut['marketsize'] < 0):
        df_pred_fut.loc[(df_pred_fut['marketsize']<0),'marketsize'] = 0
        
    df_roll = pd.concat([df_full,df_pred_fut])
    wd = order[indmin]
    df_model = df_roll.rolling(window=wd).mean()
    df_pred = df_model.iloc[-3:]#CAmodel.predict(start=CA_test.index[0], end=CA_test.index[-1])
   
    model_df.loc[model_df['state']==state,'pred_ma'] = df_pred.values
```

Example output of the relative error and optimal parameterization for each model per state is shown below. This dataframe will be used to select the best model. For example, in the case
of Alabama, the Holt-Winters model (parameterized with a seasonal_period = 3) is the
most favorable model and is expected to provide the most accurate forecast of Alabama's future market size.

```python
err_df.head()
```

Next I select the predicted market size from the optimal model for each state and combine those values into a new dataframe.

```python
honey_pred_future = pd.DataFrame(columns=['state','year','marketsize'])
for state in st_mdl_list:
    err_ar = np.asarray([err_df[err_df['state']==state][['err_lin','err_ma','err_hw']].values])
    code = np.argmin(err_ar)
    if code == 0:
        prednum = model_df[model_df['state']==state][['year','pred_lin']]
        prednum.rename(columns = {'pred_lin':'marketsize'}, inplace = True)
    elif code == 1:
        prednum = model_df[model_df['state']==state][['year','pred_ma']]
        prednum.rename(columns = {'pred_ma':'marketsize'}, inplace = True)
    elif code == 2:
        prednum = model_df[model_df['state']==state][['year','pred_hw']]
        prednum.rename(columns = {'pred_hw':'marketsize'}, inplace = True)
    
    prednum['state'] = state
    honey_pred_future = pd.concat([honey_pred_future,prednum],sort=True)
```

# Predicting the Price of Honey with Machine Learning

Now that I have a forecast for the future market size of honey in each
state, I can construct a model to use that information to predict the future price. To do that I am going to utilize 
machine learning regression algorithms. I start with a linear model and then construct a pipeline to explore polynomial models. Last, I consider a Random Forest regression model. 

```python
#import relevant machine learning libraries
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

First step is to select the features I want to model and split the dataset into a training and testing set. 

```python
X = honey_tot[honey_tot['year']<2015][['stnum','marketsize','year']]
y = honey_tot[honey_tot['year']<2015]['priceperlb']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
X = honey_tot[honey_tot['year']<2015][['stnum','marketsize','year']]
y = honey_tot[honey_tot['year']<2015]['priceperlb']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
linear = LinearRegression()
linear.fit(X_train, y_train)
linear.score(X_train, y_train)
```

```python
pred_lin=linear.predict(X_test)
```

```python
print 'RMSE: ', np.sqrt(metrics.mean_squared_error(y_test,pred_lin))
```

The root-mean-squared error and, in particular, the training score indicate the linear model can be improved upon. Next I build a pipeline for considering a range of polynomial models. The results of a cross validation routine in the pipeline are shown in the validation curve below and used deterine the optimal polynomial model. 

```python
# import libraries needed to do polynomial regression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import validation_curve
```

```python
def PolynomialRegression(degree=2, **kwargs):
    return make_pipeline(PolynomialFeatures(degree),
                         LinearRegression(**kwargs))

def rms_error(model, X, y):
    y_pred = model.predict(X)
    return np.sqrt(np.mean((y - y_pred) ** 2))

degree = np.arange(0, 6)
val_train, val_test = validation_curve(PolynomialRegression(), X, y,
                                       'polynomialfeatures__degree', degree, cv=7,
                                       scoring=rms_error)
```

```python
def plot_with_err(x, data, **kwargs):
    mu, std = data.mean(1), data.std(1)
    lines = plt.plot(x, mu, '-', **kwargs)
    plt.fill_between(x, mu - std, mu + std, edgecolor='none',
                     facecolor=lines[0].get_color(), alpha=0.2)

plt.figure(figsize=(8, 6))
plot_with_err(degree, val_train, label='training scores')
plot_with_err(degree, val_test, label='validation scores')
plt.xlabel('degree',fontsize=16); 
plt.ylabel('RMS Error',fontsize=16)
plt.ylim(0,1)
plt.xlim(0,5)
plt.legend();
```

According to the validation curve above, the best model is a degree=3 polynomial. The root-mean-squared error is minimized at degree=3 and the validation and training scores are not significantly different, so the model is not overfitting the data.

```python
tri_model = PolynomialRegression(3)
tri_model.fit(X_train, y_train)
print tri_model.score(X_train, y_train)
pred_tri=tri_model.predict(X_test)
```

The 3rd degree polynomial regressions model has an improved goodness of fit from the linear case. However, this trainging score indicates that there is some non-linearity in this model that may be better fit with a different algorithm. I explore a Random Forest regression model next. 

Random Forests are an ensemble method that utilize a collection of Decision
Trees built from subsamples of the dataset. A Decision Tree is a machine
learning algorithm that makes a series of decisions on how to informatively
divide the training data in order to predict the target feature. Random
forests make accurate predictions by aggregating the results of a large
number individual trees. This method excels in identifying non-linear
relationships and is robust against overfitting due to its internal
calculation of error during training. This makes Random Forest regression an ideal choice for the problem at hand.

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
forrest = RandomForestRegressor(n_estimators=600)
forrest.fit(X_train, y_train)
forrest.score(X_train, y_train)
```

```python
pred_rf=forrest.predict(X_test)
print 'RMSE: ', np.sqrt(metrics.mean_squared_error(y_test,pred_rf))
```

The training score and root-mean-squared error indictate that the Random Forest model is out performing all other models. As a final check I compare the prediction accuracy for the 3rd degree polynomial and Random Forest models below. 

```python
fig, ax = plt.subplots(nrows=2,ncols=3,figsize=(14, 10))
#-- Comparing the test price to predictions and plotting the residuals for the Random Forest model
sns.regplot(x=y_test,y=pred_rf, ax=ax[0,0])
sns.residplot(x=y_test,y=pred_rf,lowess=True, ax=ax[0,1])
sns.distplot(y_test-pred_rf,bins=40,axlabel='Residual',ax=ax[0,2])
ax[0,0].set_ylabel('Predicted price per lb (USD)',fontsize=14)
ax[0,0].set_xlabel('Actual price per lb (USD)',fontsize=14)
ax[0,0].set_ylim(0,3.5)
ax[0,0].set_xlim(0,3.5)
ax[0,1].set_ylabel('residual',fontsize=14)
ax[0,1].set_xlabel('Actual price per lb (USD)',fontsize=14)
ax[0,1].set_ylim(-1.5,1.5)
ax[0,2].set_xlim(-2,2)
#-- Comparing the test price to predictions and plotting the residuals for the 3rd degree polynomial model
sns.regplot(x=y_test,y=pred_tri, ax=ax[1,0])
sns.residplot(x=y_test,y=pred_tri,lowess=True, ax=ax[1,1])
sns.distplot(y_test-pred_tri,bins=40,axlabel='Residual',ax=ax[1,2])
ax[1,0].set_ylabel('Predicted price per lb (USD)',fontsize=14)
ax[1,0].set_xlabel('Actual price per lb (USD)',fontsize=14)
ax[1,0].set_ylim(0,3.5)
ax[1,0].set_xlim(0,3.5)
ax[1,1].set_ylabel('residual',fontsize=14)
ax[1,1].set_xlabel('Actual price per lb (USD)',fontsize=14)
ax[1,1].set_ylim(-1.5,1.5)
ax[1,2].set_xlim(-2,2)

plt.tight_layout();
```

These plots show the Random Forest model is making more accurate predictions than the optimal polynomial case. The distribution of the predicted and actual price have a nearly one-to-one trend for both models and the distribution of the residuals for both models have an approximately normal distribution. Both cases have a small trend in the residual dsitribution relative to larger values of the actual price, however the small number points there make that trend less concerning. Comparing the top row to the bottom, the Random Forest model is shown to be more accurate evidenced by the narrower distribution of residuals.
